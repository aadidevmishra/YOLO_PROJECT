```python
from ultralytics import YOLO
import gradio as gr
import cv2
import numpy as np
from PIL import Image
```

```python
model_fast = YOLO("yolov8n.pt")   # Fast
model_accurate = YOLO("yolov8m.pt")  # Accurate
```

```python
def detect_objects(image, yolo_version):
    img = np.array(image)

    if yolo_version == "YOLOv8 Fast":
        results = model_fast(img)
        pros = "‚úÖ Very fast\n‚úÖ Low memory usage"
        cons = "‚ùå Less accurate on small objects"
    else:
        results = model_accurate(img)
        pros = "‚úÖ High accuracy\n‚úÖ Better object localization"
        cons = "‚ùå Slower\n‚ùå More GPU usage"

    annotated_img = results[0].plot()

    summary = f"""
    üîç **{yolo_version} Analysis**

    **Advantages:**
    {pros}

    **Disadvantages:**
    {cons}
    """

    return Image.fromarray(annotated_img), summary
```

```python
with gr.Blocks(css=css) as demo:
    gr.Markdown("# üöÄ YOLO Vibe Detection Project")

    with gr.Row():
        image_input = gr.Image(type="pil", label="üì∑ Upload Image")
        image_output = gr.Image(label="üß† Detection Output")

    yolo_choice = gr.Radio(
        ["YOLOv8 Fast", "YOLOv8 Accurate"],
        label="Choose YOLO Version",
        value="YOLOv8 Fast"
    )

    text_output = gr.Markdown()

    detect_btn = gr.Button("üî• Detect Objects")

    detect_btn.click(
        detect_objects,
        inputs=[image_input, yolo_choice],
        outputs=[image_output, text_output]
    )

demo.launch(share=True)
```
